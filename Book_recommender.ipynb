```python
import pandas as pd
```

```python
books = pd.read_csv("Books.csv")
```

```python
books
```

```python
books.rename(columns ={ 'Book-Title':'title',"Book-Author":"author","Year-Of-Publication":"year","Publisher":"publisher"},inplace = True)
```

```python
books.head()
```

```python
users = pd.read_csv("Users.csv")
ratings = pd.read_csv("Ratings.csv")
```

```python
users.rename(columns ={'User-ID':'user_id', 'Location':'location', 'Age':'age'},inplace=True)
ratings.rename(columns = {'User-ID':'user_id', 'Book-Rating':'rating'},inplace =True)
```

```python
users.tail
```

```python
ratings.head()
```

```python
X=ratings['user_id'].value_counts()
```

```python
no_of_user_counted = ratings['user_id'].value_counts()>200
users_counted = X[no_of_user_counted].index
```

```python
users_counted
```

```python
ratings = ratings[ratings['user_id'].isin(users_counted)]
```

```python
ratings.shape
```

```python
merged_books = ratings.merge(books,on='ISBN')
```

```python
merged_books
```

```python
times_rating_given = merged_books.groupby('title')['rating'].count().reset_index()
times_rating_given.rename(columns ={ 'rating':'number_of_rating'
},inplace= True)
```

```python
times_rating_given.tail()
```

```python
 books_with_final_rating= merged_books.merge(times_rating_given, on='title')
```

```python
books_with_final_rating
```

```python
books_with_final_rating = books_with_final_rating[books_with_final_rating['number_of_rating'] >=50]
```

```python
books_with_final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
books_with_final_rating.shape
```

```python
book_pivot = books_with_final_rating.pivot_table(columns = 'user_id',index ='title',values='rating')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distances,sugggestions = model.kneighbors(book_pivot.iloc[741,:].values.reshape(1,-1))
```

```python
distances,sugggestions
```
